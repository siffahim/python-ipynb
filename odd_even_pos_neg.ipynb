```python
n = int(input());
inp = input();
numbers = inp.split();

odd = 0;
even = 0;
pos = 0;
neg = 0;

for i in range(n):
    num = int(numbers[i]);

    if num < 0:
        neg += 1;
    if num > 0:
        pos += 1;
    
    if num%2 == 0:
        even += 1;
    else:
        odd += 1;



print("Even:", even);
print("Odd:", odd);
print("Positive:", pos);
print("Negative:", neg);
```
