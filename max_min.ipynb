```python
inpt = input();
numbers = inpt.split();

a = int(numbers[0]);
b = int(numbers[1])
c = int(numbers[2])

#finding max
max_n = 0;
if a > b and a > c:
    max_n = a;
elif b > a and b > c:
    max_n = b;
else:
    max_n = c;


#finding min
min_n = 0;
if a < b and a < c:
    min_n = a;
elif b < a and b < c:
    min_n = b;
else:
    min_n = c;

print(min_n, max_n);
```
