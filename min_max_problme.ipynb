```python
inpt = input();
numbers = inpt.split();

a = int(numbers[0]);
b = int(numbers[1])
c = int(numbers[2])

#finding max
max_n = a;
min_n = a;

if b < min_n:
    min_n = b;
if c < min_n:
    min_n = c;


#finding min
if b > max_n:
    max_n = b;
if c > max_n:
    max_n = c;

print(min_n, max_n);
```
